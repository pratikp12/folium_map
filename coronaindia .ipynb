```python
import folium
from folium.features import DivIcon
import pandas as pd
data=pd.read_csv("D:\indian_state_maps\corona.csv")

lon=list(data["Long"])
lat=list(data["Lat"])
no=list(data["CNFMD"])

map = folium.Map(location=[21, 78], zoom_start=5)

#fgv=folium.FeatureGroup(name="Corona_virus")


for lt,ln,n in zip(lat,lon,no):
    folium.Marker([lt,ln],icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(7,10),
        html='<div style="font-size: 20t; color:red">'+str(n)+'</div>',
        )).add_to(map)
    map.add_child(folium.CircleMarker([lt,ln], radius=12))
    
    
map

```

```python
import folium
import pandas as pd

state_geo = 'states2.json'


state_data = pd.read_csv('corona1.csv')

#Let Folium determine the scale
map = folium.Map(location=[21, 78], zoom_start=5)

folium.Choropleth(geo_data=state_geo, data=state_data,
         columns=['STATE', 'CNFMD'],
         key_on='feature.id',
         fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.2,legend_name='Cases').add_to(map)

folium.LayerControl().add_to(map)
map

```
